```python
import pandas as pd
import numpy as np
from scipy.signal import argrelextrema

df = pd.read_csv("AAPL.csv")
```

```python
def get_minmax(e):
    minimum = np.asarray(argrelextrema(e, np.less, order=2))
    maximum = np.asarray(argrelextrema(e, np.greater, order=2))
    return minimum, maximum
```

```python
c=get_minmax(e)
print(c)
```

```python
e[c[0]], e[c[1]] # access index's values here
```

```python
c_merged = np.concatenate(maximum, minimum)
c_merged = np.array(c_merged)
print(c_merged)
```
